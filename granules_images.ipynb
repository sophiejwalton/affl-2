```python
def get_fnames(folder, date, genotype, n, HS = True ):
    print(folder)
    fnames = []
    for i in range(n):
        i += 1
        
            
        if i < 10:
            st_i = '0' + str(i)
        else:
            st_i = str(i)
        if HS:
            fnames.append(folder + '/' + date + '_' + genotype + '_' + '100%int100ms480_05umZ_HS' + '-' +  st_i
                      + '.tif')
        else:
            fnames.append(folder + '/' + date + '_' + genotype + '_' + '100%int100ms480_05umZ_NOHS' + '-' + st_i
                      + '.tif')
    print(fnames)
    return fnames

def get_stuff(folder, genotype, dates, HS, nums = [10,10], blobs = False, background = True):
    fnames = []
    dates_l = []
    for i, d in enumerate(dates):
        dates_l = dates_l + nums[i]*[d]
    print(dates_l)
    for n, d in zip(nums, dates):
        fnames = fnames + get_fnames(folder, d, genotype, n, HS = HS)
        dates_l = dates_l + [len(get_fnames(folder, d, genotype, n, HS = HS))]
  
    ic = []
    labeled = []
    filt = []
    dfs = []
    for i, f in enumerate(fnames):
        im_labeled, im_filt, im, df = full_process(f, blobs = blobs, background = background)
        
        ic.append(im)
        labeled.append(im_labeled)
        filt.append(im_filt)
        if len(df) > 5: 
            
            dfs.append(df)
            df['date'] = dates_l[i]
            df['worm id'] = i + 1
    full_df = pd.concat(dfs)
    full_df['genotype'] = genotype
    return full_df, ic, labeled, filt
```

```python
def full_process(fname, selem = skimage.morphology.disk(10), selem2 = skimage.morphology.disk(2), 
                 std = 40.0, min_size = 400, threshR = 1., threshS = 1800, blobs = False, background = True):
    print(fname)
    im = load_img(fname)
    

    try:
        im_black = load_img(fname[:-4] + '_black.tif')
    except: 
        im_black = skimage.io.imread(fname[:-4] + '_black.tif')
    _, im_snp_filt = granule_filt(im, std = std)
    im_bw, _ = granule_thresh(im_snp_filt, selem=selem,
                   min_size=min_size, selem2 = selem2)
    im_labeled, im_filt, df = get_nuclei(im_bw, im, im_black, threshR = threshR, threshS = threshS, blobs = blobs,
                                        background = True )
    return im_labeled, im_filt, im, df
```
